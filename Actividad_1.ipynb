```python
##2.1
# Estandarización de datos
marca = "Aguila"

if marca.lower() == "budweiser":
    print("compro", marca) # Más recomendado
elif marca.lower() == "aguila":
    print("compra", marca)
else:
    print("No compro nada")
```
